The history of NLP can be divided into five major stages.

---

## Stage 1: Bag-of-Words (BoW)

### Idea

A method for representing unstructured text by counting word occurrences.

### How it works:
1. **Tokenization:** Split sentences into individual words/tokens
2. **Vocabulary Creation:** Combine all unique words
3. **Counting:** Count how often each word appears
4. **Vector Representation:** Create numerical vectors

Example:

Sentence:

```
My cat is cute
```

Vocabulary:

```
my
cat
is
cute
dog
```

Vector:

| my | cat | is | cute | dog |
| -- | --- | -- | ---- | --- |
| 1  | 1   | 1  | 1    | 0   |

---

### Advantages

* Simple
* Fast
* Interpretable

---

### Limitations

Ignores:

* Word order
* Context
* Meaning

Example:

```
Dog bites man
```

and

```
Man bites dog
```

produce similar vectors.

---

````python
# Code Example: Bag-of-Words Implementation
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Sample sentences
sentences = [
    "I love llamas",
    "Ik hou van lama's"  # Dutch translation
]

# Create bag-of-words representation
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(sentences)

# Display results
print("Vocabulary:", vectorizer.get_feature_names_out())
print("\nBag-of-Words Matrix:")
df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
print(df)
````



## Stage 2: Word2Vec (2013)

Major breakthrough. Instead of counting words:

Represent words as dense vectors.

### Key Features:
- **Dense vectors:** Fixed-size arrays (e.g., 50-300 dimensions)
- **Semantic similarity:** Similar words have similar vectors
- **Context-based:** Learns from word co-occurrence patterns
- **Training:** Uses neural networks to predict word relationships

---

### Key Insight

Words appearing in similar contexts have similar meanings.

Example:

```
King
Queen
Prince
Princess
```

end up near each other in vector space.

---

### Famous Analogy

```
King - Man + Woman ≈ Queen
```

This emerges naturally from embeddings.

# 3. Embeddings

## Definition:

Embeddings are dense numerical representations of language.

Instead of:

```
Cat → [0,0,1,0,0,0]
```

we get:

```
Cat → [0.23, -0.91, 0.45, ...]
```

---

## Why Embeddings Matter

Embeddings capture:

* Meaning
* Context
* Similarity

---

### Example

```
Cat
Dog
Puppy
```

are close together.

```
Apple
Banana
```

are close together.

---

## Types of Embeddings

### Word Embeddings

Represent a word.

Example:

```
cat
dog
house
```

---

### Sentence Embeddings

Represent an entire sentence.

Example:

```
My cat is cute.
```

---

### Document Embeddings

Represent entire documents.

Useful for:

* Search
* Retrieval
* RAG

---



````python
# Code Example: Working with Word Embeddings
# Note: This requires gensim library
# !pip install gensim

try:
    from gensim.models import Word2Vec
    import numpy as np

    # Sample training data (tokenized sentences)
    sentences = [
        ["i", "love", "llamas"],
        ["llamas", "are", "amazing", "animals"],
        ["i", "enjoy", "animals"],
        ["babies", "are", "cute"]
    ]

    # Train word2vec model
    model = Word2Vec(sentences, vector_size=50, window=2, min_count=1, workers=4)

    # Get embedding for a word
    word = "llamas"
    embedding = model.wv[word]
    print(f"Embedding for '{word}':")
    print(f"Shape: {embedding.shape}")
    print(f"First 10 values: {embedding[:10]}")

    # Find similar words
    print(f"\nWords similar to 'llamas':")
    similar = model.wv.most_similar("llamas", topn=2)
    for word, score in similar:
        print(f"  {word}: {score:.4f}")

except ImportError:
    print("Install gensim to run this example: !pip install gensim")
````

# 8. Transformers

Paper:

### "Attention Is All You Need" (2017)

Changed AI forever.

---

## Main Idea

Remove RNNs. Use attention everywhere.

Allows models to focus on relevant parts of the input sequence when processing or generating text.

### Key Ideas:
- **Selective focus:** Model "attends" to important words
- **Context encoding:** Captures relationships between words
- **Parallel processing:** Can process entire sequence at once
- **Weighted importance:** Different words get different attention scores
---

## Benefits:

Solves the problem of long sequences losing context in RNNs.

### Parallel Processing

Old:

```
Word 1
↓
Word 2
↓
Word 3
```

Transformer:

```
Word1 Word2 Word3
processed simultaneously
```

---

### Faster Training

Much faster than RNNs.

---

### Better Context

Can see the entire sequence.

---

# 9. Self-Attention

The core of Transformers.

---

## Question

How important is every word to every other word?

Example:

```
I love llamas
```

Self-attention computes:

|        | I    | love   | llamas |
| ------ | ---- | ------ | ------ |
| I      | High | Low    | Low    |
| love   | Low  | High   | Medium |
| llamas | Low  | Medium | High   |

---

The model learns relationships automatically.

---

# 10. Transformer Architecture

Two main components.

---

## Encoder

Purpose:

```
Understand language
```
#### **Encoder Block:**
1. Self-attention layer
2. Feedforward neural network
3. Layer normalization
4. Residual connections


---

## Decoder

Purpose:

```
Generate language
```

#### **Decoder Block:**
1. Masked self-attention (only attends to previous tokens)
2. Cross-attention (attends to encoder output)
3. Feedforward neural network
4. Layer normalization

### Key Features:
- **Stacked blocks:** Multiple encoder/decoder layers
- **Parallel training:** Much faster than RNNs
- **Autoregressive:** Generates one token at a time
- **Positional embeddings:** Adds sequence order information
---

# 11. BERT (Encoder-Only)

Bidirectional Encoder Representations from Transformers (2018).
---

## Purpose

Understand text.

Not generate text.

---

### Architecture:
- Uses only encoder blocks
- Bidirectional context (looks at entire sequence)
- Special [CLS] token for classification tasks

### Training Method: Masked Language Modeling
- Masks random tokens in input
- Model predicts masked tokens
- Creates deep contextual understanding


Example:

```
I love [MASK]
```

Predict:

```
llamas
```

---

## Applications

* Classification
* Search
* Clustering
* NER
* Sentiment Analysis

---

**Key Advantage:** Excellent at understanding and representing language context.

````python
# Code Example: Using BERT for Text Representation
# Note: This requires transformers library
# !pip install transformers torch

try:
    from transformers import AutoTokenizer, AutoModel
    import torch

    # Load BERT model and tokenizer
    model_name = "bert-base-uncased"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name)

    # Sample text
    text = "Large language models are powerful tools for natural language processing"

    # Tokenize and get model output
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

    # Get embeddings
    with torch.no_grad():
        outputs = model(**inputs)

    # Extract [CLS] token embedding (first token)
    cls_embedding = outputs.last_hidden_state[:, 0, :]

    print(f"Text: {text}")
    print(f"CLS embedding shape: {cls_embedding.shape}")
    print(f"Embedding dimension: {cls_embedding.shape[1]}")

except ImportError:
    print("Install transformers and torch: !pip install transformers torch")
````

# 12. Decoder-Only Models (GPT)

**Model:** Generative Pre-trained Transformer (2018)

**Purpose**: Generate text.

### Architecture:
- Uses only decoder blocks
- Unidirectional (left-to-right)
- Autoregressive generation
- Masked self-attention (prevents looking ahead)

### Evolution:
- **GPT-1:** 117M parameters (7,000 books)
- **GPT-2:** 1.5B parameters
- **GPT-3:** 175B parameters
- **GPT-3.5:** Powers ChatGPT
- **GPT-4:** Most advanced (proprietary)

### Training Paradigm:

#### **Step 1: Pretraining (Language Modeling)**
- Train on vast internet text corpus
- Learn grammar, context, patterns
- Predict next word
- Result: Foundation/Base model
- Very expensive (e.g., Llama 2: 2T tokens, $5M+)

#### **Step 2: Fine-tuning**
- Further train on specific tasks
- Adapt to desired behavior
- Less compute-intensive
- Examples: Instruction following, classification


### Training Objective

Predict next token.

Example:

Input:

```
I love
```

Output:

```
llamas
```

---

Then:

Input:

```
I love llamas
```

Output:

```
because
```

and so on.

---

## Autoregressive Generation

Each new token depends on previous tokens.

# 14. Context Window

**Context Length/Window:**
- Maximum number of tokens model can process
- Increases as new tokens are generated
- Larger windows = more context

**Completion Models:**
- Take input text and autocomplete it
- Can be fine-tuned for chat/instruction following



````python
# Code Example: Understanding Model Parameters and Context
import sys

# Calculate approximate memory requirements
def calculate_model_memory(num_parameters, precision=4):
    """
    Calculate memory required for a model

    Args:
        num_parameters: Number of model parameters
        precision: Bytes per parameter (4=float32, 2=float16, 1=int8)

    Returns:
        Memory in GB
    """
    memory_bytes = num_parameters * precision
    memory_gb = memory_bytes / (1024**3)
    return memory_gb

# GPT model sizes
models = {
    "GPT-1": 117_000_000,
    "GPT-2": 1_500_000_000,
    "GPT-3": 175_000_000_000,
    "Phi-3-mini": 3_800_000_000
}

print("Model Memory Requirements (float32):")
print("-" * 40)
for name, params in models.items():
    memory = calculate_model_memory(params, precision=4)
    print(f"{name:15} {params:>12,} params → {memory:6.2f} GB")

print("\nModel Memory Requirements (float16):")
print("-" * 40)
for name, params in models.items():
    memory = calculate_model_memory(params, precision=2)
    print(f"{name:15} {params:>12,} params → {memory:6.2f} GB")
````

````python
# Code Example: Comparing Model Sizes and Requirements
import math

def model_requirements(params_billion, context_length=4096):
    """
    Estimate model requirements
    """
    # Memory in different precisions (GB)
    fp32 = params_billion * 4 / 1024
    fp16 = params_billion * 2 / 1024
    int8 = params_billion * 1 / 1024
    int4 = params_billion * 0.5 / 1024

    # KV cache memory (approximate)
    # Formula: 2 * layers * hidden_size * context_length * precision
    # Simplified estimate
    kv_cache_gb = (params_billion * context_length) / (1024 * 1024) * 0.1

    return {
        "fp32": fp32,
        "fp16": fp16,
        "int8": int8,
        "int4": int4,
        "kv_cache": kv_cache_gb
    }

# Analyze different models
models_info = {
    "Phi-3-mini": 3.8,
    "Llama-3-8B": 8,
    "Mistral-7B": 7,
    "Llama-2-13B": 13,
    "Llama-2-70B": 70
}

print("Model Memory Requirements by Precision (GB)")
print("=" * 70)
print(f"{'Model':<20} {'FP32':>8} {'FP16':>8} {'INT8':>8} {'INT4':>8} {'Total*':>8}")
print("-" * 70)

for model, params in models_info.items():
    reqs = model_requirements(params)
    total = reqs["int4"] + reqs["kv_cache"]  # Typical setup
    print(f"{model:<20} {reqs['fp32']:>8.2f} {reqs['fp16']:>8.2f} {reqs['int8']:>8.2f} {reqs['int4']:>8.2f} {total:>8.2f}")

print("\n*Total = INT4 weights + KV cache (typical local setup)")
print("\nRecommendation: Use INT4 quantization for local deployment!")
````
